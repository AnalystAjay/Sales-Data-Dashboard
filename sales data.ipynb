```python
## Sales Data Analysis
import pandas as pd 

df = pd.read_csv('sales_data_clean_utf8.csv')
```

```python
#Check the first few rows of the data
df.head()
```

```python
#Get summary statistics
df.info()
```

```python
# Describe the sales column
df.describe()
```

```python
# check for missing values
df.isnull().sum()
```

```python
# Fill missing values in ADDRESSLINE2 with 'Unknown'
df['ADDRESSLINE2'] = df['ADDRESSLINE2'].fillna('Unknown')
```

```python
# Fill missing values in STATE with 'Unknown'
df['STATE'] = df['STATE'].fillna('Unknown')
```

```python
# Fill missing values in POSTALCODE with the mode
df['POSTALCODE'] = df['POSTALCODE'].fillna(df['POSTALCODE'].mode()[0])
```

```python
# Fill missing values in TERRITORY with 'Unknown'
df['TERRITORY'] = df['TERRITORY'].fillna('Unknown')
```

```python
# check again for missing values after filling 'unknown' and mode 
df.isnull().sum()
```

```python
# Check the last few rows of the data
df.tail(10)
```

```python
# Standardize column names to lowercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
# Check the updated column names
df.columns
```

```python
# Check for duplicate rows
df.duplicated().sum()
```

```python
# View duplicate rows if any
df[df.duplicated()]
```

```python
# Final check of the data structure and types
df.info()
```

```python
# Convert orderdate to datetime format
df['orderdate'] = pd.to_datetime(df['orderdate'], errors='coerce')
```

```python
# Convert sales, quantityordered, and priceeach to numeric types
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
df['quantityordered'] = pd.to_numeric(df['quantityordered'], errors='coerce')
df['priceeach'] = pd.to_numeric(df['priceeach'], errors='coerce')
```

```python
# Convert categorical columns to category data type
df['status'] = df['status'].astype('category')
df['productline'] = df['productline'].astype('category')
df['country'] = df['country'].astype('category')
```

```python
#Check for missing values
df.isnull().sum()
```

```python
# Check the updated column names
df.columns
```

```python
# Create a new column 'full_address' by combining 'addressline1' and 'addressline2'
df['full_address'] = (
    df['addressline1'].fillna('') + ', ' + df['addressline2'].fillna('')
).str.strip(', ')
```

```python
# Create a new column 'contact_fullname' by combining 'contactfirstname' and 'contactlastname'
df['contact_fullname'] = df['contactfirstname'] + ' ' + df['contactlastname']
```

```python
# Remove outliers from the sales column using the IQR method
Q1 = df['sales'].quantile(0.25)
Q3 = df['sales'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['sales'] >= Q1 - 1.5*IQR) & (df['sales'] <= Q3 + 1.5*IQR)]
```

```python
#Date feature engineering - extract year, month, and day from orderdate
df = df.copy()
df['orderdate'] = pd.to_datetime(df['orderdate'], errors='coerce')

# Extract date parts
df['year'] = df['orderdate'].dt.year
df['month'] = df['orderdate'].dt.month
df['day'] = df['orderdate'].dt.day
```

```python
df.columns
```

```python
df['month_name'] = df['orderdate'].dt.strftime('%b')
```

```python
month_map = {
    1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr',
    5:'May', 6:'Jun', 7:'Jul', 8:'Aug',
    9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'
}

df['month_name'] = df['month_id'].map(month_map)

df.to_csv('sales_data_final.csv', index=False)
```

```python
#check the first few rows of the updated data
df.head(5)
```

```python
# Create a new column 'estimated_profit' by calculating 20% of the sales
df = df.copy()
df['estimated_profit'] = df['sales'] * 0.2
```

```python
# Group by productline and calculate total estimated profit for each product line
df.groupby('productline')['estimated_profit'].sum()
```

```python
# Create a new column 'sales_category' based on sales amount
df = df.copy()
df['sales_category'] = pd.cut(df['sales'],
                             bins=[0,1000,5000,10000,50000],
                             labels=['low','medium','high','very_high'])
```

```python
# Calculate total sales, total orders, and total customers
print("Total Sales:", df['sales'].sum())
print("Total Orders:", df['ordernumber'].nunique())
print("Total Customers:", df['customername'].nunique())
```

```python
# Group by country and calculate total sales for each country, then sort in descending order
df.groupby('country')['sales'].sum().sort_values(ascending=False)
```

```python
# Group by customername and calculate total sales for each customer, then sort in descending order and show top 10 customers
df.groupby('customername')['sales'].sum().sort_values(ascending=False).head(10)
```

```python
# Group by productline and calculate total sales and total quantity ordered for each product line, then sort by total sales in descending order
df.groupby('productline').agg({
    'sales':'sum',
    'quantityordered':'sum'
}).sort_values(by='sales', ascending=False)
```

```python
# Group by year and month, and calculate total sales for each month
df.groupby(['year','month'])['sales'].sum()
```

```python
# Group by status and calculate total sales for each status
df.groupby('status')['sales'].sum()
```

```python
# Create a pivot table to show total sales for each product line in each country
pd.pivot_table(df,
               values='sales',
               index='country',
               columns='productline',
               aggfunc='sum')
```

```python
# Calculate the correlation between sales, quantityordered, and priceeach
df[['sales','quantityordered','priceeach']].corr()
```

```python
# RFM Analysis - Calculate Recency, Frequency, and Monetary value for each customer
import datetime as dt

latest_date = df['orderdate'].max()

rfm = df.groupby('customername').agg({
    'orderdate': lambda x: (latest_date - x.max()).days,
    'ordernumber': 'count',
    'sales': 'sum'
})

rfm.columns = ['recency','frequency','monetary']
print(rfm.head())
```

```python
# Visualize total sales by product line using a bar chart
import matplotlib.pyplot as plt
df.groupby('productline')['sales'].sum().plot(kind='bar')
plt.title("Sales by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Visualize monthly sales trend using a line chart
monthly_sales = df.groupby(['year', 'month'])['sales'].sum()

monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# Visualize sales distribution by status using a bar chart
df.groupby('country')['sales'].sum().plot(kind='bar')
plt.title("Sales by Country")
plt.ylabel('Percentage of Total Sales')
plt.show()
```

```python
# Visualize top 10 customers by total sales using a bar chart
top_customers = df.groupby('customername')['sales'].sum().sort_values(ascending=False).head(10)

top_customers.plot(kind='bar')
plt.title("Top 10 Customers")
plt.show()
```

```python
# Visualize sales distribution using a histogram
Q1 = df['sales'].quantile(0.25)
Q3 = df['sales'].quantile(0.75)
IQR = Q3 - Q1

filtered_df = df[(df['sales'] >= Q1 - 1.5*IQR) & (df['sales'] <= Q3 + 1.5*IQR)]

filtered_df['sales'].plot(kind='hist', bins=20)
plt.title("Sales Distribution ")
plt.show()
```

```python
# Visualize the relationship between price and sales using a scatter plot
sample_df = df.sample(500)

plt.scatter(
    sample_df['priceeach'], 
    sample_df['sales'], 
    s=sample_df['quantityordered']*5,
    alpha=0.5
)

plt.xlabel("Price")
plt.ylabel("Sales")
plt.title("Price vs Sales ")
plt.show()
```

```python
# Visualize weekly sales trend using a line chart
df['week'] = df['orderdate'].dt.isocalendar().week

df.groupby('week')['sales'].sum().plot(kind='line')
plt.title("Weekly Sales Trend")
plt.show()
```
